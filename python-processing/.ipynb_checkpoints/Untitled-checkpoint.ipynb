```python
import json
import networkx as nx

# Load the JSON file
with open("../graph-data-files/graph_data-3.json") as f:
    data = json.load(f)

# Initialize a graph
G = nx.Graph()

# Dictionary to map node IDs to (lat, lng) and labels
node_positions = {}
labeled_nodes = {}  # Store labeled nodes

# Add nodes
for node in data["nodes"]:
    node_id = node["id"]
    latlng = (node["latlng"]["lat"], node["latlng"]["lng"])  # Tuple of (lat, lng)
    G.add_node(node_id, lat=latlng[0], lng=latlng[1], color=node["color"], label=node["label"])
    
    # Store node positions
    node_positions[node_id] = latlng
    
    # Collect labeled nodes
    if node["label"]:  
        labeled_nodes[node_id] = node["label"]

# Add edges
for edge in data["edges"]:
    G.add_edge(edge["node1"], edge["node2"], id=edge["id"])

# Compute shortest paths between all labeled nodes
shortest_paths = {}
for src_id in labeled_nodes:
    shortest_paths[src_id] = {}
    for dest_id in labeled_nodes:
        if src_id != dest_id:
            try:
                # Get the shortest path as a list of node IDs
                path = nx.shortest_path(G, source=src_id, target=dest_id)
                
                # Convert node IDs to (lat, lng) tuples
                path_coords = [node_positions[node] for node in path]
                
                shortest_paths[src_id][dest_id] = path_coords
            except nx.NetworkXNoPath:
                shortest_paths[src_id][dest_id] = None  # No path exists

# Print shortest paths with lat/lng instead of node IDs
for src, paths in shortest_paths.items():
    print(f"Shortest paths from {labeled_nodes[src]} ({node_positions[src]}):")
    for dest, path in paths.items():
        if path:
            print(f"   {labeled_nodes[dest]} ({node_positions[dest]}): {path}")
            print()
        else:
            print(f"   {labeled_nodes[dest]} ({node_positions[dest]}): No path found")
    print('-------------------------------------------------------------------')
```

<h3>This makes sure all nodes are connected</h3>

```python
import json
import networkx as nx

def clean_graph(input_file, output_file):
    # Load the JSON file
    with open(input_file, "r") as f:
        data = json.load(f)

    # Initialize a graph
    G = nx.Graph()

    # Store node details in a dictionary
    node_ids = {node["id"]: node for node in data["nodes"]}

    # Add nodes
    for node in data["nodes"]:
        G.add_node(node["id"])

    # Add edges
    for edge in data["edges"]:
        G.add_edge(edge["node1"], edge["node2"])

    # Get the first node in the JSON file
    if node_ids:
        first_node = next(iter(node_ids))  # Get the first node's ID
    else:
        raise ValueError("No nodes in the JSON file.")

    # Find all nodes connected to the first node
    if first_node in G:
        connected_nodes = set(nx.node_connected_component(G, first_node))
    else:
        connected_nodes = set()

    # Find unconnected nodes
    unconnected_nodes = set(node_ids.keys()) - connected_nodes

    # Find nodes that are directly connected to unconnected nodes
    directly_connected_to_removed = set()
    for node in unconnected_nodes:
        if node in G:
            directly_connected_to_removed.update(G.neighbors(node))

    # Remove both unconnected nodes and their direct neighbors
    nodes_to_remove = unconnected_nodes | directly_connected_to_removed
    G.remove_nodes_from(nodes_to_remove)

    # Create the cleaned JSON data
    cleaned_nodes = [node for node in data["nodes"] if node["id"] in G.nodes]
    cleaned_edges = [edge for edge in data["edges"] if edge["node1"] in G.nodes and edge["node2"] in G.nodes]

    cleaned_data = {
        "nodes": cleaned_nodes,
        "edges": cleaned_edges
    }

    # Save the cleaned JSON in the same format
    with open(output_file, "w") as f:
        json.dump(cleaned_data, f, indent=4)

    print(f"Cleaned graph saved to {output_file}")

# Example usage:
input_file = "../graph-data-files/graph_data-3.json"
output_file = "cleaned_graph_data.json"
clean_graph(input_file, output_file)
```
